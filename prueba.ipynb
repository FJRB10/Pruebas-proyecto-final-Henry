```python
def agrupacion_ciudades(ciudad:str, df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11):
    dict_ciudad = {}
    lista_ciudades = []
    lista_sector = []
    lista_valor = []
    lista_df = [df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11]

    for df in lista_df:
        lista_index = df[df['address'].str.contains(ciudad, case = False, na = False)]['category'].value_counts().index
        lista_values = df[df['address'].str.contains(ciudad, case = False, na = False)]['category'].value_counts().values

        for i in range(5):
            lista_ciudades.append(ciudad)
            lista_sector.append(lista_index[i][0])
            lista_valor.append(lista_values[i])

    dict_ciudad['Ciudad'] = lista_ciudades
    dict_ciudad['Sector'] = lista_sector
    dict_ciudad['Valor'] = lista_valor

    df_ciudad = pd.DataFrame(dict_ciudad)

    df_sum_ciudad = df_ciudad.groupby('Sector')['Valor'].sum().reset_index().sort_values(by = 'Valor', ascending = False).reset_index(drop = True)

    return pd.DataFrame(df_sum_ciudad)
```

```python
df_LA = agrupacion_ciudades("Los Angeles", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_NY = agrupacion_ciudades("New York", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_CHI = agrupacion_ciudades("Chicago", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_HOU = agrupacion_ciudades("Houston", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_PHX = agrupacion_ciudades("Phoenix", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_PHL = agrupacion_ciudades("Philadelphia", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_TUS = agrupacion_ciudades("Tucson", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_TPA = agrupacion_ciudades("Tampa", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_IND = agrupacion_ciudades("Indianapolis", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
df_BNA = agrupacion_ciudades("Nashville", df_google, df_google2, df_google3, df_google4, df_google5, df_google6, df_google7, df_google8, df_google9, df_google10, df_google11)
```
